```python
import numpy as np
np.set_printoptions(precision=4)

import matplotlib.pyplot as plt

def LoadData(name):
    f = open(name)
    ret = []
    for line in f:
        ret.append(np.array([ float(i) for i in line.split() ]))
    return ret

data = []
data.append(LoadData("m1.txt"))
data.append(LoadData("m2.txt"))
data.append(LoadData("m3.txt"))
data.append(LoadData("m4.txt"))
data = np.array(data)

for i in data:
    print(*i)
```

```python
mean = []
for row in data:
    mean.append([])
    for vals in row:
        mean[-1].append(np.mean(vals))
mean = np.array(mean)
        
for row in mean:
    print(row)
```

```python
##8
mCarriage    = 0.047  ##Kilos
errMCarriage = 0.0005 ##Kilos
mPuck        = 0.220  ##Kilos
errMPuck     = 0.0005 ##Kilos
mWeights     = 0.408  ##Kilos
errMWeights  = 0.0005 ##Kilos
r0           = 0.057  ##Meters
rStep        = 0.025  ##Meters
height       = 0.700  ##Meters
errHeight    = 0.001  ##Meters
dHub         = 0.046  ##Meters
errDHub      = 0.0005 ##Meters
g            = 9.806  ##Meters*second^-2
lenWeight    = 0.040  ##Meters

masses = np.array([ mCarriage + mPuck * i for i in range(1, 5) ])
radiuses = np.array([ r0 + rStep * i + lenWeight * 0.5 for i in range(6) ])

print(masses)
print(radiuses)
```

```python
errMean = []
for i in range(len(data)):
    errMean.append([])
    for j in range(len(radiuses)):
        errMean[-1].append(np.std(data[i][j]) / np.sqrt(len(data[i][j])))
errMean = np.array(errMean)
        
print(errMean)
```

```python
acc = (2 * height) / (mean ** 2)
print(acc)
```

```python
eps = (2 * acc) / dHub
print(eps)
```

```python
torque = []
for i in range(len(masses)):
    torque.append([])
    for j in range(len(radiuses)):
        torque[-1].append(0.5 * masses[i] * dHub * (g - acc[i][j]))
torque = np.array(torque)
print(torque)
```

```python
errAcc = np.sqrt((2 * errHeight / height) ** 2 + (2 * errMean / mean) ** 2) * acc
errEps = np.sqrt((2 * errAcc / acc) ** 2 + (errDHub / dHub) ** 2) * eps
errMasses = np.sqrt((errMCarriage / mCarriage) ** 2 + (np.array([1, 2, 3, 4]) * errMPuck / mPuck) ** 2) * masses

errTorque = []
for i in range(len(masses)):
    errTorque.append(np.sqrt((errMasses[i] / masses[i]) ** 2
                             + (errHeight / height) ** 2 + (errAcc[i] / acc[i]) ** 2) * torque[i])
errTorque = np.array(errTorque)
    
print(errAcc, errEps, errTorque, sep='\n\n')
```

```python
def LeastSquares(xs, ys):
    A = np.array([ [i, 1] for i in xs ])
    return np.linalg.lstsq(A, ys, rcond=None)[0]

inertia = [0] * len(radiuses)
friction = [0] * len(radiuses)

for i in range(len(radiuses)):
    xs = eps[:, i]
    ys = torque[:, i]
    inertia[i], friction[i] = LeastSquares(xs, ys)
    print(np.array([radiuses[i], inertia[i], friction[i]]))
```

```python
MOfEps = lambda x, idx: inertia[idx] * x + friction[idx]

fig, axs = plt.subplots(2, 3, figsize=(25, 16))

for i in range(len(radiuses)):
    xs = eps[:, i]
    ys = torque[:, i]
    errXs = errEps[:, i]
    errYs = errTorque[:, i]
    axs[int(i / 3), i % 3].errorbar(x=xs, y=ys, xerr=errXs, yerr=errYs, fmt='c', ecolor='r', label="errors")
    axs[int(i / 3), i % 3].plot([xs[0], xs[-1]], [MOfEps(xs[0], i), MOfEps(xs[-1], i)], 'm', label="fitted line")
    axs[int(i / 3), i % 3].legend()
```

```python
radiusesSq = radiuses ** 2
M0, I0 = LeastSquares(radiusesSq, inertia)

errRadiusesSq = []
errInertia = []

K = -1 / M0
for (x, y) in zip(radiusesSq, inertia):
    b = y - K * x
    x0 = (I0 - b) / (K - M0)
    y0 = M0 * x0 + I0
    errRadiusesSq.append(np.abs(x - x0))
    errInertia.append(np.abs(y - y0))

IOfRSq = lambda x: M0 * x + I0

fig, axs = plt.subplots(1, figsize=(10, 6))
axs.errorbar(x=radiusesSq, y=inertia, xerr=errRadiusesSq, yerr=errInertia, fmt='c', ecolor='r', label="errors")
axs.plot([radiusesSq[0], radiusesSq[-1]], [IOfRSq(radiusesSq[0]), IOfRSq(radiusesSq[-1])], 'm', label="fitted line")
axs.legend()
plt.show()
```

```python
print("%.5f" % np.std(errRadiusesSq))
print("%.5f" % np.std(errInertia))
```
