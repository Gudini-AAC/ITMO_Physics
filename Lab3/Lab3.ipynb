```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from scipy import interpolate
```

```python
def load_data(number_of_lines, path):
    file = open(path)
    
    lines = []
    for line in file:
        lines.append(line)
    
    potentials = []
    ret = []
    
    for i in range(number_of_lines):
        potentials.append(float(lines[4 * i]))
        ys = [ float(j) for j in lines[4 * i + 1].split() ]
        xs = [ float(j) for j in lines[4 * i + 2].split() ]
        ret.append(np.array([xs, ys]))
    
    return potentials, ret
```

```python
has_circle = False

path = ''
count = 0
if has_circle:
    path = 'data_2.txt'
    count = 13
else:
    path = 'data_1.txt'
    count = 9

a, b = load_data(count, path)
    
def to_arrays(potentials, lines):
    xs = []
    ys = []
    zs = []
    
    for i in range(len(lines)):
        number_of_points = len(lines[i][0])

        for x in lines[i][0]:
            xs.append(x)
        for y in lines[i][1]:
            ys.append(y)
        for j in range(number_of_points):
            zs.append(potentials[i])
    
    return xs, ys, zs

xs, ys, zs = to_arrays(a, b)


if has_circle:
    angles = np.linspace(0, np.pi * 2, 50)

    for angle in angles:
        xs.append(15 + np.cos(angle) * 5)
        ys.append(10 + np.sin(angle) * 5)
        zs.append(7.18)
        
tck = interpolate.bisplrep(xs, ys, zs)
```

```python
grid_spacing = .1 # cm
full_x = np.arange(2, 28, grid_spacing)
full_y = np.arange(2, 18, grid_spacing)

mesh = np.array(np.meshgrid(full_x, full_y))

full_V = np.transpose(interpolate.bisplev(full_x, full_y, tck))
```

```python
E = -np.array(np.gradient(full_V))
E_norm = np.hypot(E[0], E[1])

E = E / E_norm

if has_circle:
    mask = np.hypot(mesh[0] - 15, mesh[1] - 10) < 5
    E = np.where(mask, 0, E)
    E_norm = np.where(mask, 0, E_norm)
```

```python
plt.rcParams["figure.figsize"] = (16, 8)

max_v = np.max(full_V)
min_v = np.min(full_V)

cmap = cm.get_cmap('viridis')

cp = plt.contourf(full_x, full_y, full_V, 100, cmap=cmap)
cb = plt.colorbar(cp, label='Potential φ, volts')

plt.streamplot(full_x, full_y, E[1], E[0], color='red')

if has_circle:
    circle_color = cmap((7.18 - min_v) / (max_v - min_v))
    plt.gca().add_patch(plt.Circle((15, 10), 5.0, ec='black', color=circle_color, zorder=10))
    plt.gca().add_patch(plt.Circle((15, 10), 4.0, ec='black', color=circle_color, zorder=10))

plt.show()
```

```python
cp = plt.contourf(full_x, full_y, E_norm / (grid_spacing * 0.01), 100, cmap=cmap)
cb = plt.colorbar(cp, label='Electric field strength, volts/meter')

plt.streamplot(full_x, full_y, E[1], E[0], color='red')

if has_circle:    
    circle_color = cmap(0.05)
    plt.gca().add_patch(plt.Circle((15, 10), 5.0, ec='black', color=circle_color, zorder=10))
    plt.gca().add_patch(plt.Circle((15, 10), 4.0, ec='black', color=circle_color, zorder=10))

plt.show()
```

```python
linear_xs = full_x
linear_ys = full_V[80]

plt.plot(linear_xs, linear_ys, label='First case: φ(x, 10), volts')
plt.plot(circle_xs, circle_ys, label='Second case: φ(x, 10), volts')
plt.minorticks_on()
plt.grid(which="major", color="k", linewidth=0.5, alpha=0.5)
plt.grid(which="minor", color="k", linestyle=":", linewidth=0.5, alpha=0.5)
plt.legend()
plt.show()
```
